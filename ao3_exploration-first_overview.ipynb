#### User actions on AO3

##### 1. Comments 
Comments are a method of leaving feedback on a work. Each work, or each chapter of a work, has a textbox at the end where you can input your remarks. You can also respond to comments other people have left, which will form a comment thread. Comments are located at the bottom of a work's page, beneath the list of kudos. Note that a work can have multiple pages of comments; each page will show 20 comment threads. Comment threads are like conversations: one person leaves a comment, and all subsequent replies to that comment are part of one comment thread. For example, if somebody comments on one of your works saying, "Wow! This is great, have you published anything else on the Archive?" and you reply, "Yes! Here's a link to one of my other works!", the original comment and all its replies are one comment thread. If you respond to other comments, your replies will be directly below the comments you respond to. These will be indented to show they are replies from the same thread.

By default, you can comment anonymously on any work in the Archive; however, the work creator can also disable this function if they choose. After you delete your account, any comments you've left with that account will be attributed to the username "Account Deleted." When you change your username, any comments and/or kudos you've left under your old username will update to your new username. When you leave kudos while accessing a particular chapter of a work, it's linked to the work as a whole. 

##### 2. Kudos

The word "kudos" is from the ancient Greek, meaning "glory" or "renown". One modern definition is "praise given for achievement". As an Archive feature, kudos are a quick and easy way to let a creator know that you like their work. You can only give kudos to a work once, even if you have multiple pseuds under your account. Kudos cannot be deleted at this time, either by the user who gave kudos or by the user who received it. After you delete your account, any kudos you've left with that account will be attributed to a guest rather than to your username.

##### 3. Bookmarks

An AO3 bookmark is a record of a work you want to remember, find more easily, or record a note about. Bookmarks created on the Archive can also serve as recommendations to other users. Any work or series posted on the Archive can be bookmarked. You can also add your own personal notes or tags for the work. The notes can be anything, such as brief reviews or reminders to yourself. You can also mark a bookmark as a Rec (Recommendation). If you mark a bookmark as a Rec, it will be included in search results when a user searches for Recs. When viewing bookmarks, any that have been marked as a Rec will display with a heart icon on the blurb. Unlike kudos, bookmarks can be edited and/or removed.

#### Obtaining the data

As previewed in the initial section, the data that is used in this notebook was obtained by scraping the AO3 platform. The content of AO3 is organized in fandoms. Each fandom reprents a community of users built around a real world work such as a movie, a video game, a book or a TV series. When there are multiple media types available for the same title - Harry Potter, for example, has multiple books and movies released within the same universe, the fandom doesn't make a distiction between these two formats. 


The scraper is based on a repo made publicly available on github (https://github.com/ArmindoFlores/ao3_api) and it was modified to better suit large scale scraping. One of the most important improvements that were made was optimizing how comments were retrieved in order to minimize the bandwidth use and the number of requests. Since comments are structured in threads, one might retrieve all top-level comments on a page, and then recursively inspect the immediate child threads that each top-level thread contains. 

By default, an AO3 comment page loads a certain number of comments for each thread, and in case the conversation is very long, the thread will get clipped and a "Show more" button will be added. This button will load new comments in that thread. Most of the threads are not longer than 2 or 3 comments, so more often than not there will be no hidden comments. As a consequence, these threads can be considered fully displayed when the page is loaded. This means that on average, one requests to the AO3 servers is the equivalent to loading a page of comments. 

There are some exceptions, however, when there is the need to make further requests  in order to load all the comments on a page. There requests correponds to the threads that have to be loaded in order to read the  deeply-nested comments. Even in this cases, it is sufficient to load the last comment in the chain, the one before the "Show more button". As an example, if there were five comments, labeled A to E, each replying to the previous one (i.e. B replies to A and so on) and by default only the first 3 comments of a thread are shown, this means that the comments from A to C will be visible, while comments D and E will be hidden. 

picture

The original approach implies that in order to load all the comments, we need to load all the threads. The term thread is used to refer to a conversation made of multiple comments. Based on what comment we choose as a root, we might find multiple threads for such a conversation. In the previous example, there will be a thread rooted in A, that will contain a thread rooted in B and so on. But since the comments B and C are already displayed when the page is loaded, we can directly load the thread rooted in C in order to read the other comments down the chain, while the thread rooted in B would give us only a new comment (D), instead of two.  This new approach makes comment scraping much faster and more efficient in terms of network resources. 

AO3 gives the possibility to select a fandom and browse its works. Works are further structered in chapters and they are not necessarily released all at once. Users often publish their fanfics with only a few chapters, and later add new ones down the road. It's up to the author to choose when a work can be considered as complete and mark it as such. Each page contains 20 *works* - the specific terms used on AO3 to refer to fanfics - and for each of them a preview is shown containing the most basic metadata for that work. This preview includes the following items:
- the username of the author who has written the fanfic
- the title of the fanfic
- the tags that the author of the fanfic has chosen to associate to their work
- the date when the fanfic was posted
- a short plot 
- the language of the fanfic
- the number of words, chapters, kudos, bookmarks, comments and hits (or views)
Some of the details are not explicitly displayed, but they can be retrieved anyway from the HTML attributes. Such examples are the work ID, and the links to comments, bookmarks and kudos, although we need only the first one. Using the work ID, we can infer the others by building the URL manually (e.g: the bookmarks of the work with ID equal to x, will have its bookmarks at the endpoint /works/x/bookmarks).


picture

This data can therefore be scraped by iterating through the pages of each fandom we are interested in, and each page will offer the metadata for 20 works, without any information about the single chapters, apart their number.

So far we have described how the works metadata and the comments are retrieved. Obtaining kudos and bookmarks represents an easier task compared to comments since these are listed as lists of items, without any nesting among items. Bookmarks are always loaded in batches of 20 per page and are described by an ID, the username of the user who bookmarked the work, the timestamp of the bookmark and finally an optional comment. Kudos are even simpler since they are only a list of the users who offered their appreciation through them and they are loaded in batches of 50.

It is worth noticing that different interactions (i.e. comments, bookmarks,kudos) offer different kind of attributes. Kudos are the most common but they are attached to a work and have no timestamp information. On the other side, Comments are the least common, belong to a specific chapter inside a work, and have the date and time when they were posted. Bookmarks are situated somewhere between the two in terms of how common and how rich their attributes are. Similarly to comments, they have a timestamp, but just as kudos, each bookmark is connected to a work rather than a chapter.

#### Storing the data

In the previous sections we focused on the data mainly from a semantic point of view. This part covers the technical aspects of how the data is structured and stored on the permanent storage. We now know that the instances used to store and manipulate the data are supported by MongoDB. Each fandom has its data stored across two *collections*. a MongoDB collection can be seen as the equivalent of the tables in relational databases. The first collection contains  all the metadata described in the previous section, together with the bookmarks and the kudos, while the second one is dedicated to the comments. Thus, the database contains a total of 14 collection , split across 7 collections for the comments and 7 collections for the remaining data. Collections are then further split into documents. If we wante to reuse the same analogy as earlier, the documents of a NoSQL database would correspond to the rows in a SQL database table. Documents however are less rigid and allow to build more complex schemas. In fact, it's possible to structure data into nested items in order to create a sort of data tree represntation.

Let's look into detail at one document example for the comment collection and another one for the metadata, kudos and bookmarks. Please note that the choice of the fandom is not relevant since they are all using the same field number and names.

##### Example of Harry Potter collection document:

- _id (ObjectID): This ID is the default key that is generated by Mongo to uniquely identify each document in the collection. Its type is a special Mongo type  and it consists of 12-byte hexadecimal string values
- id: the work ID on AO3 (integer)
- authors(array of strings): a list of users that have contributed to writing the work. Most of the times the works are released under a single user name, although there are a few exceptions (less than 1% of the works in each fandom) where there are collaborations on producing the fanfic.
- title (string): title of the work.
- nchapters (integer): the number of chapters that have been released so far (integer)
- expected_chapters (integer or string): the number of chapters that the author is expecting to release. It is often the case that the author does not know this number, so it will appear as a question mark on AO3. The Mongo DB documents capture this information by storing the "N/A" placeholder value which will make this field become a string rather than an integer.
- complete (boolean): indicates whether the author is going to publish new chapters in the future or the work has been completed and marked as such (boolean)
- status(string): similar to the previous one, it describe the status of the work as a human readable string. The possible values are In Progress or Complete.
- date_updated (datetime): the last time the author has edited their work, stored using the ISO Date format.
- hits (integer): the number of clicks the work has received. This number increases each time the work is loaded, therefore it  might not correlate perfectly with the number of distinct users that access a specific work.
- language (string): the (main) language the work is written in, stored as human readable string (e.g: English rather than EN).
- summary (string): a short work summary written by the author of the work.
- tags (array of strings): a list of (optional) and user-defined (i.e: a new tag can be created at anytime even if it was not used on the platform before) tags added by the author (e.g: Mystery, Crossover, Drama).
- nkudos (integer or string): the number of kudos the work has received. Sometimes this number is not available on AO3, so in order to distinguish between an unknown value and an actual zero, the "N/A" value is used like we did for the number of expected chapters, hence the field becomes a string.

#### Cleaning the data

Normally the scraped data come unstructured and requires extensive cleaning procedures. For example, one would expect that information of each bookmark would be embedded inside a single row of a database, and eventually the information would be extracted from this raw data source. The good news is that the scraper did already most of the work during the data collection process by storing the data in a structured way and assigning appropriate data types to each field. This makes the stored data ready for the analysis even without further cleaning. In fact, in the initial data exploration phase,  the data was used as-is and no pre-processing was needed until later in the analysis. However, it was found that some small cleaning methods were useful in order to improve the accuracy of the results. This was also possible by getting more familiar with the domain of the data and feature specific to the platform where the data was obtained from. 

The first pre-processsing operation that was applied regards the chapter number in the documents of the collections containing comments. This value has found to be often "N/A" (string type). This is not a Mongo specific or reserved value, but a convenient way I chose to represent the lack of or an unknown value on AO3. Using the zero value of a data type, such as the 0 value for integers could have been another alternative, although it leaves more space for interpretation and is potentially ambiguos for those who will have to use the dataset in the feature. The presence of a chapter numbered 0 means that the chapters are zero-indexed or that the value is missing? Luckily, Mongo is not as strict as traditional relational databases when it comes to the data type of a field. Therefore, the chapter number can be present in a single collection as both integer and string.

The "N/A" values are common enough to skew the data analysis and they can't be mitigated by attempting to reload the page or the content since the value won't change and the issue goes beyond the efforts of the parser. What we can do instead is reduce the amount of such unknown values and infer their value where this is possible. The first step in this cleaning procedure involves retrieving the works that have only one chapter published: the logic here is that if those works have received any comments, they will necessary belong to that chapter since it is the only one. We can safely change the value of the chapter field from "N/A" to 1. This operation is done one time only by running a script that changes all the occurences that match this criterion. Later, when the dataset is loaded for analysis purposes, the data will have the pro-processed values. For the other works, whhere the number of published chapters is greater than one, we cannot infer this so the value will remain "N/A".

Another finding that improved especially the aggreggation operations was a feature that AO3 offers to the users. A single registered users can create multiple profiles and appear under different identities on the platform, by managing one password only. In the AO3 slang, these identities are called _pseuds_, an abbreviations from the word pseudonyms. For example, an user called U might have an extra two extra pseudonyms, namely P1 and P2. When publishing a work or writing a comment on another's user work, the user U can choose the identity they want to appear with. Likewise most of the plaforms that support registered users, the page of the user can be accessed by clicking the username of the user. Let's assume user U published a new work in the Harry Potter fandom under the P1 identity, so the users will see that the author of this new work is somebody called P1. They will eventually visit the user page of P1 and this action will forward them to the main profile page of user U. Here they will see a list of pseudonyms that user U owns and the activity connected to each of them.

In our analysis, grouping the activity of a user with multiple identities is more important than focusing on the individual contributions. In the end we are talking about the same person and this is a more natural unit of aggreggation rather than the platform-specific feature for handling multiple identities. Luckily, at the time when the data was collected, the scraper stored both the pseudonym of the user and the main accout name as the username of the user that performed some action. Such fields can be found in the following cases:
    - the author field in the top-level of the colllection describing the meta data of the work
    - the author field nested as a field of the bookmarks array
    - one single item of the array containing the kudos left on a work
    - the author field (i.e. the author of the comment) in the collection describing the comments.
    
The format used to store both the pseudonym and the main user name is by concatenating the first one followed by the second one between parenthesis. By building on the previous example, the author of the work would be store in the database as P1 (U). In case users chose to use their main account name, and this is normally the case on AO3 since pseuds are not so common, then the parenthesis would not be needed and a single string can be used. Pre-processing the pseudonyms involves firsly retrieving all the fields listed above that contain an opening and a closing parentheses. From there, we can split the string based on these two caracters and assign the new value of the fields to be the content enclosed between them.

As a last step in the data cleaning process, we will look at how dates are handled. Working with time and dates in general is not an easy task since the formats are virtually unlimited and parsing and normalizing them requires to be aware of all the formats that are present o the platform. Luckily, after reviewing thoroughly the values used by the fields involving the time and the date, the format could be assumed to be the following:
- day of the week as a three letter abbreviation (e.g: Wed for wednesday).
- day of the month as a number between 1 and 31.
- month of the year as a three letter abbreviation (e.g Dec for December)
- the year as a four digit integer
- time of the day using a 12 hour format and the "AM" and "PM" abbreviations to distinguish between after and before midday (e.g: 1:57PM means 13:57, while 1:57 AM meaning 1:57 in the morning).
- the timezone of the user as an abbreviation made of up to 5 letters. (e.g: CEST stands for Central European Summer Time).

Listing all the possible abbreviations is out of the scope of this description, especially in the case of timezones where the possible values are close to 100 possible options. The good news is that the date fields follow this format when present (otherwise marked with "N/A" for the reasons desribede earlier), while the bad news is that they have to use the same time zone in order to be able to compare, order and group them. The choice of the specific timezone to convert all the values to is not as important as long as the derived date and time values are all using the same one. For example, choosing the Australian Eastern Day Time (AEDT) doesn't have any tangible impact on the results later in the analysis when compared to another option such as the Central European Summer Time as long as they are consistent. 

Python offers a rich ecosystem of well-tested and publicly-available packages, so performing this kind of data pre-processing is a lot easier than in other programming languages. The fields that are going to be affected by this operation are the following:
- the date_updated and date_published fields in the collection containing the metadata of the works
- the date field of the bookmarks nested object in the collection describing the works
- the date field of the comment documents

Furthermore, having a consistent format and semantic of the date fields is not enough. Some more complex database queries such as grouping by day or month require can potentially be slow if the operation on the dates are not made efficient. As a consequence, the value of the date cannot be of type string, as this would slow the overall performance of the queries, and eventually the execution of the whole computational tasks. Mongo uses an international time and date format for representing these sort of values and has a dedicated data type when storing values that are represented according to this format. By using the standard Python package "datetime" and the "pytz" package from the Python Package Index (PyPI), it is possible to convert all the date fields to this format and with an unique time zone. The script that does this outputs the dates in a format that has the date (YYYY-MM-DD) followed by a 'T' character that separates the date from the time in 'HH:MM.MS' format where MS are the miliseconds; finally, the +TZ (timezone) is appended. This is an example: 2013-12-21T12:07:17.000+00:00. Since the chosen timezone was UTC, the final part will also equal to 00:00.

A connection to the MongoDB database is created and the names of the collections (i.e. fandoms) are specified.
Works and stories, respectively, are the names used for the fan-written fictions. Therefore, the *works_collections* list will contain the collections for the AO3 fandoms, while *stories_collections* those of FFN.
Two different tuples are used for selecting the fields we are interested in. All the documents in the two categories (i.e. databases or platforms) have the same structure.
Finally, we import the data by iterating on the collections and creating a dataframe for each of them.
Some data cleaning is done by striping the leading and trailing spaces on string field, replacing NA values with 0 and removing works whose author is either anonymous, belong to a deleted account or an archivist account. AO3 offers also the possibility to users to have extra nicknames on top of their main username, called *pseuds*. When we do aggregated operations on the users, we need all users to appear with their main username. In the databasem this name is stored inside parenthesis in case the user used a *pseud* on the website. We therefore extract the content of the parentheses and drop the rest.

### Analyzing text

**How does LIWC work?**

For example, if LIWC were analyzing the first line of the novel Paul Clifford by Edward Bulwer-Lytton (1842):

>It was a dark and stormy night

the program would first look at the word “it” and then see if “it” was in the dictionary.
It is and is coded as a function word, a pronoun, and, more specifically, an impersonal pronoun. All three of these LIWC categories would then be incremented. Next, the word “was” would be checked and would be found to be associated with the categories of verbs, auxiliary verbs, and past tense verbs.
After going through all the words in the novel, LIWC would calculate the percentage of each LIWC category. So, for example, we might discover that 2.34% of all the words in a given book were impersonal pronouns and 3.33% were auxiliary verbs. The LIWC output, then, lists all LIWC categories and the rates that each category was used in the given text.

````python
def tokenize(text):
    #It is possible to use a smarter tokenizer
    for match in re.finditer(r'\w+', text, re.UNICODE):
        yield match.group(0)
````

- Regarding the specific regex expression **r'\w+'**, this searches for groups that have alphanumerics (that's the \w part) or apostrophes (which is also in those square brackets) that are 1 or longer. Note that whitespace is not a match, so this, generally speaking, breaks a line into words.

- match_object.group(0) says that the whole part of match_object is chosen.

- yield is a keyword that is used like return, except the function will return a generator.

## Communities compared: mean, std, and variance

Remember that the _standard deviation_ (std) tells us, on average, how far each score lies from the mean in your data set, while the _variance_ tells us the degree of spread in the data set. The more spread the data, the larger the variance is in relation to the mean.

_It doesn't make much sense to put categories side by side, also since one word might repeat in multiple categories_. **You have to take a category, like Posemo and compare it between different communities**.

We will focus on the following major categories of emotions...
- **Emotional expression** – _positive_ and _negative emotion_ | _anxiety_, _anger_, _sadness_
- **Other language features** – _personal pronouns_ like first, second, third person | _singular_, _plural_ | _male_ vs _female references_
- **Perception process** – _see_, _hear_, _feel_

They map on LIWC's categories as follows:
- **Emotional expression** – _Posemo_ and _Negemo_ | _Anx_, _Anger_, _Sad_
- **Other language features** – _Ppron_ like first, second, third person | first-person singular *I* and third-person singular _SheHe_, first-person plural _We_ and third-person plural _They_, second-person _You_, impersonal _Ipron_ | _Male_ vs _Female_
- **Perception process** – _See_, _Hear_, _Feel_

````python
liwc_categories_main = [ 'Posemo', 'Negemo', 'Anx', 'Anger', 'Sad', 'Ppron', 'I', 'SheHe', 'We', 'They', 'You', 
                        'Ipron', 'Male', 'Female', 'See', 'Hear', 'Feel' ]
labels = FANDOMS.copy()
````

````python
def plot_liwc_categories_mean_std_variance(category):
    x = np.arange(1, len(labels) + 1)
    y = np.array([])
    e = np.array([])
    for fandom in FANDOMS:
        y = np.append(y, works_collected_categories_desc[fandom][category][1])
        e = np.append(e, works_collected_categories_desc[fandom][category][2])
    f = lambda e: e**2
    variance = f(e)
    #Plotting the std
    f, ax = plt.subplots(figsize=(16, 8))
    f.subplots_adjust(right=.65)
    plt.errorbar(x, y, e, linestyle='None', fmt='o', capsize=3, capthick=3)
    plt.xticks(x, FANDOMS_PRETTY)
    plt.xticks(rotation=45)
    for k, g in enumerate(range(1, len(labels) + 1)):
        v = y[k] + e[k]
        r = e[k]
        ax.annotate(f'$\sigma$  {r:.2f}', (g, v), textcoords='offset points',
                    xytext=(0, 3), ha='center', va='bottom', fontsize='x-small')
        ax.annotate(f'{y[k]:.2f}', (g+.17, y[k]-.15), textcoords='offset points',
                    xytext=(0, 3), ha='center', va='bottom', fontsize='x-small')
    plt.title(f'{category} mean and std')
    plt.ylabel('Mean and std')
    plt.show()
    #Plotting the variance
    f, ax = plt.subplots(figsize=(16, 8))
    f.subplots_adjust(right=.65)
    plt.errorbar(x, y, variance, linestyle='None', fmt='o', capsize=3, capthick=3)
    plt.xticks(x, FANDOMS_PRETTY)
    plt.xticks(rotation=45)
    for k, g in enumerate(range(1, len(labels) + 1)):
        v = y[k] + e[k]**2
        r = e[k]**2
        ax.annotate(f'$\sigma$\u00b2  {r:.2f}', (g, v), textcoords='offset points',
                    xytext=(0, 3), ha='center', va='bottom', fontsize='x-small')
        ax.annotate(f'{y[k]:.2f}', (g+.17, y[k]-.15), textcoords='offset points',
                    xytext=(0, 3), ha='center', va='bottom', fontsize='x-small')
    plt.title(f'{category} mean and variance')
    plt.ylabel('Mean and variance')
    plt.show()
    return y, e, variance
````
